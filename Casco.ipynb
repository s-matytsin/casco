```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
casco_train = pd.read_csv('C:\\Users\\monte\\Downloads\\casco_train.csv\\casco_train.csv')
casco_train
```

```python
casco_train = casco_train.rename(columns = {'column_1':'Gender', 'column_2':'Drive_Experience', 'column_3':'Marital_Status',
                                            'column_4':'City_Of_Driver_Registration','column_5':'City_of_Contract_registration',
                                            'column_6':'Maximum_Payments_Delay', 'column_7':'Engine_Type',
                                            'column_8':'Vehicle_Mileage', 'column_9':'Vehicle_Manufacturing_Year',
                                            'column_10':'Engine_Power', 'column_11':'Number_Of_Client_Contracts',
                                            'column_12':'Number_Of_Vehicle_Contracts', 'column_13':'Age', 'column_14':'Vehicle_Type',
                                            'column_15':'Vehicle_Manufacturing_Country','column_16':'Telematics_Systems_Presence',
                                            'column_17':'Brand_And_Model','column_18':'Vehicle_Technical_Parameter',})
```

```python
casco_train
```

```python
casco_train = casco_train.drop_duplicates()
```

```python
casco_train.Drive_Experience = casco_train.Drive_Experience.fillna(casco_train.Drive_Experience.median())
```

```python
casco_train.loc[casco_train.Vehicle_Mileage > 220000, 'Vehicle_Mileage'] = casco_train.Vehicle_Mileage.median()
```

```python
casco_train = casco_train.drop(['City_of_Contract_registration',
                                'City_Of_Driver_Registration','Gender'], axis = 1)
```

```python
from sklearn import preprocessing
LE = preprocessing.LabelEncoder()
```

```python
casco_train['Marital_Status'] = LE.fit_transform(casco_train['Marital_Status'])
casco_train['Engine_Type'] = LE.fit_transform(casco_train['Engine_Type'])
casco_train['Vehicle_Manufacturing_Country'] = LE.fit_transform(casco_train['Vehicle_Manufacturing_Country'])
casco_train['Brand_And_Model'] = LE.fit_transform(casco_train['Brand_And_Model'])
```

```python
X_train = casco_train.drop('target', axis = 1)
y_train = casco_train.target
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
DTR = DecisionTreeRegressor(max_depth = 10,max_leaf_nodes = 512, random_state = 123)
regressor = AdaBoostRegressor(DTR, n_estimators = 1000, learning_rate = 0.025)
```

```python
regressor.fit(X_train,y_train)
```

```python
regressor.score(X_train,y_train)
```

```python
regressor.feature_importances_
```

```python
importances = pd.Series(regressor.feature_importances_, index = X_train.columns)
importances.sort_values(ascending=False)
```
